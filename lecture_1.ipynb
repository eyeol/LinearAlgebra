### 0. Intro.
선형대수는 본질적으로 연립일차방정식의 Solution을 구하는 학문이다.

일반적으로 연립방정식은 n개의 방정식에 n개의 변수가 주어지는데,<br>
연립일차방정식을 바라보는 관점(Picture)에 따라 수학적인 해석이 달라진다.
<br>
<br>

### 1. Row Picture and Column Picture
<br>
$
2x - y = 0\\
-x + 2y = 3
$
<br>
먼저 2개의 방정식과 2개의 변수가 있는 연립방정식을 Row Picture(행의 관점)으로 해석해보자.

```python
x=np.arange(-6,8)

y = 2*x
y2 = (1/2)*x + 3


plt.plot(x,y, 'b')
plt.plot(x,y2, 'r')


plt.show()
```
